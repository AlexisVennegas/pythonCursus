```python

## como primer paso importamos las librerias que vamos a utilizar
import re 

cadena = "Este un es un ejemplo de una cadena de caracteres"

## distintos metodos



## print(libreria(re).metodo(search)argumentos(argument1, argument2)
##el primer argumento es lo que queremos buscar y el segundo es la cadena de caracteres donde queremos buscar
##esto devuelve un objeto de tipo match que es un objeto que contiene informacion sobre la busqueda
##si no encuentra nada devuelve None
print("---- search ----")
print(re.search("cadena", cadena)) 

##resultado = <re.Match object; span=(29, 35), match='cadena'>
## (re.match) == objeto de tipo match devuelto por el search
## (span) == devuelve la posicion donde se encuentra la palabra buscada
## (match) == devuelve el texto que se encontro

print("---- match ----")
print(re.search("hoda", cadena)) 
## resultado = None no devuelve ni false ni true 
## no se encontro la palabra buscada
 
textoBuscar = "ejemplo"
print("---- findall ----")
if re.search(textoBuscar, cadena) is not None:
    print("Se encontro la palabra buscada")
else:
    print("No se encontro la palabra buscada")




textoEncontrado = re.search(textoBuscar, cadena)


print("--- another example --- ")
print(textoEncontrado.start()) ## devuelve la posicion donde se encuentra la palabra buscada
print(textoEncontrado.end()) ## devuelve la posicion donde termina la palabra buscada

if(textoEncontrado):
    print("funciona")
else:
    print("no funciona")

print("---- split ----")
## divide la cadena de caracteres en una lista de subcadenas
print(re.split(" ", cadena)) ## devuelve una lista de subcadenas

print("---- sub ----")
## reemplaza la palabra buscada por otra
print(re.sub("ejemplo", "otra", cadena)) ## devuelve la cadena de caracteres con la palabra reemplazada


print("---- finditer ----")
## devuelve un iterador con los objetos match de todas las coincidencias
for textoEncontrado in re.finditer("un", cadena):
    print(textoEncontrado.start(), textoEncontrado.end())

print("---- span ----")
## devuelve una tupla con la posicion donde se encuentra la palabra buscada
print(re.search("ejemplo", cadena).span())
```

```python
import re

## metodo compile 
## permite compilar una expresion regular para poder reutilizarla

## ejemplo buscaremos numeros en una cadena


pattern = r'\d+' ## \d+ == cualquier numero positivo o negativo o cero o float

## creamos un objeto de tipo regex
regex = re.compile(pattern)

## buscamos en una cadena de caracteres

cadena = "123456789"
cadena2 = "hola mundo"
cadena3 = "hola 123 mundo 456"

## buscamos en la cadena de caracteres

matches = regex.findall(cadena2)
print("---- compiler ----")
if matches:
    print(matches)
else:
    print("no se encontro nada")


## patrones comunmente utilizados

## \d == cualquier numero
## \D == cualquier caracter que no sea un numero
## \s == cualquier espacio en blanco
## \S == cualquier caracter que no sea un espacio en blanco
## \w == cualquier caracter alfanumerico
## \W == cualquier caracter que no sea alfanumerico
## \b == espacio en blanco al principio o al final de una palabra
## \B == no es un espacio en blanco al principio o al final de una palabra
## ^ == coincide con el inicio de una cadena de caracteres
## $ == coincide con el final de una cadena de caracteres
## [] == coincide con un conjunto de caracteres
## * == coincide con 0 o mas repeticiones del patron
## + == coincide con 1 o mas repeticiones del patron
## ? == coincide con 0 o 1 repeticiones del patron




## ejemplo de uso de los patrones


## busquemos correos

pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'
"""[A-Za-z0-9._%+-]+: Coincide con uno o más caracteres alfanuméricos, puntos, guiones bajos, porcentajes o signos de suma y resta, que forman el nombre de usuario en la dirección de correo electrónico."""
## @: Coincide con el carácter @.
"""[A-Za-z0-9.-]+: Coincide con uno o más caracteres alfanuméricos, puntos o guiones, que forman el nombre de dominio en la dirección de correo electrónico."""
## \.: Coincide con el carácter punto.
"""[A-Za-z]{2,}: Coincide con dos o más caracteres alfabéticos, que forman el dominio de nivel superior en la dirección de correo electrónico."""

text = 'Estos son algunos correos electrónicos de ejemplo: user@example.com, john.doe@gmail.com, info@company.co.uk'

regex = re.compile(pattern)

matches = regex.findall(text)

print("---- correos ----")
if matches:
    print(matches)
else:
    print("no se encontro nada")


```

```python
import re 

## ejemplo de uso de los patrones
list_name = ["Alexis Venegas", 
             "Juan Perez",
             "Maria Ana Garcia",
             "Pedro Rodriguez",
             "Luisa Martinez",
             "Jose Gonzalez",
             "Ana Lopez",
             "Maria Garcia Ana",
             "Ana Jimenez"]

## busquemos nombres que empiecen con Ana
print("---- empieza con Ana ----")
for name in list_name:
    if re.findall('^Ana', name):
        print(name)

## busquemos nombres que terminen con Ana
print("---- termina con Ana ----")
for name in list_name:
    if re.findall('Ana$', name):
        print(name)


## busquemos nombres que tengan la "ñ"

print("---- tiene ñ ----")
lista_empresas = ["http://www.ñandu.com.ar","http://www.google.com", "http://empresa-españa.com"]

for empresa in lista_empresas:
    if re.findall('ñ', empresa):
        print(empresa)



print("---patrones de busqueda--- ")

lista_nombres = ["hombres", "mujeres", "niños", "niñas", "mascotas", "animales"]

for persona in lista_nombres:
    if re.findall('niñ[oa]s', persona): ## [oa] == puede ser o ó a
        print(persona)


print("----rango de busqueda----")

lista_personas = ["hombres", "mujeres", "niños", "niñas", "mascotas", "animales"]

for persona in lista_personas:
    if re.findall('[a-z]', persona): ## [a-z] == puede ser cualquier caracter de la a a la z
        print(persona)
```
